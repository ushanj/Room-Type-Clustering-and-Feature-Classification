```python
import pandas as pd
import numpy as np
import gensim
```

```python
# Load the pre-trained word2vec model
model_path = 'GoogleNews-vectors-negative300.bin'
model = gensim.models.KeyedVectors.load_word2vec_format(model_path, binary=True)
```

```python
# Define the list of words for bed sizes
bed_size_key_words = ['Twin', 
    'Double', 
    'Queen', 
    'King' ,
    'California King',
    'Twin XL' ,
    'Full' ,
    'Single' ,
    'Sofa' ,
    'Rollaway' ,
    'Bunk' ,
    'Murphy' ,
    'Trundle' ,
    'Daybed',
    'Crib' , 'Pack' 
    ,'Air Mattress', 
    'Twin beds', 
    'Double beds', 
    'Queen beds', 
    'King beds' ,
    'California King beds',
    'Twin XL beds' ,
    'Full beds' ,
    'Single beds' ,
    'Sofa beds' ,
    'Rollaway beds' ,
    'Bunk beds' ,
    'Murphy beds' ,
    'Trundle beds' ,
    'Daybed beds',
    'Crib beds' , 'Pack beds'                  
 ]



room_style_key_words = [
    'Queen'
    'Standard',
    'Suite',
    'Deluxe',
    'Executive',
    'Club',
    'Superior',
    'Connecting',
    'Accessible',
    'Family',
    'Pet-Friendly',
    'Jacuzzi',
    'Adjoining',
    'Loft',
    'Penthouse Suite',
    'Standard Room',
    'Suite Room',
    'Deluxe Room',
    'Executive Room',
    'Club Room',
    'Superior Room',
    'Connecting Room',
    'Accessible Room',
    'Family Room',
    'Pet-Friendly Room',
    'Jacuzzi Room',
    'Adjoining Room',
    'Loft Room',
    'Penthouse Suite Room'
    
]
```

```python
df = pd.read_csv("merged.csv")
df = df.head(100)
```

```python


# Define a function to tokenize the room types and extract bigrams and trigrams
def get_ngrams(room_type):
    tokens = room_type.split()
    bigrams = [' '.join(tokens[i:i+2]) for i in range(len(tokens)-1)]
    trigrams = [' '.join(tokens[i:i+3]) for i in range(len(tokens)-2)]
    return tokens + bigrams + trigrams


# Apply the get_ngrams function to the room types column
df['ngrams'] = df['room_type'].apply(get_ngrams)

# Define a function to find the best matching key word for a list of ngrams
def get_best_match(ngrams, key_words):
    best_match = ''
    best_similarity = -1
    for ngram in ngrams:
        if ngram in model:
            for key_word in key_words:
                # Split key word into individual words
                key_word_tokens = key_word.split()
                # Check if all words in key word are present in ngram
                if all(token in ngram.split() for token in key_word_tokens):
                    similarity = model.similarity(ngram, key_word)
                    if similarity > best_similarity:
                        best_similarity = similarity
                        best_match = key_word
    return best_match

# Apply the get_best_match function to the ngrams column for bed sizes
df['best_matching_bed_size'] = df['ngrams'].apply(lambda x: get_best_match(x, bed_size_key_words))

# Apply the get_best_match function to the ngrams column for room styles
df['best_matching_room_style'] = df['ngrams'].apply(lambda x: get_best_match(x, room_style_key_words))
```
