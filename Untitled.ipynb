```python
import pandas as pd
import numpy as np
import gensim
```

```python
# Load the pre-trained word2vec model
model_path = 'GoogleNews-vectors-negative300.bin'
model = gensim.models.KeyedVectors.load_word2vec_format(model_path, binary=True)
```

```python
# Define the list of words for bed sizes
bed_size_key_words = ['Twin', 
    'Double', 
    'Queen', 
    'King' ,
    'California King',
    'Twin XL' ,
    'Full' ,
    'Single' ,
    'Sofa' ,
    'Rollaway' ,
    'Bunk' ,
    'Murphy' ,
    'Trundle' ,
    'Daybed',
    'Crib' , 'Pack' 
    ,'Air Mattress' 
 ]



room_style_key_words = [
    'Queen',
    'Queen Room'
    'Standard Room',
    'Suite',
    'Deluxe Room',
    'Executive Room',
    'Club Room',
    'Superior Room',
    'Connecting Room',
    'Accessible Room',
    'Family Room',
    'Pet-Friendly Room',
    'Jacuzzi Room',
    'Adjoining Room',
    'Loft Room',
    'Penthouse Suite'
    
]
```

```python

#tokenize the room types and extract bigrams and trigrams
def get_ngrams(room_type):
    tokens = room_type.split()
    bigrams = [' '.join(tokens[i:i+2]) for i in range(len(tokens)-1)]
    trigrams = [' '.join(tokens[i:i+3]) for i in range(len(tokens)-2)]
    return tokens + bigrams + trigrams


# Apply get_ngrams function to the room types column
df['ngrams'] = df['room_type'].apply(get_ngrams)
```

```python
df.head(1)
```

```python
def get_best_match(ngrams, key_words):
    best_match = ''
    best_similarity = -1
    for ngram in ngrams:
        if ngram in model:
            for key_word in key_words:
                # Split key word into individual words
                key_word_tokens = key_word.split()
                # Check if all words in key word are present in ngram
                if all(token in ngram.split() for token in key_word_tokens):
                    similarity = model.similarity(ngram, key_word)
                    if similarity > best_similarity:
                        best_similarity = similarity
                        best_match = key_word
    return best_match


# Apply the get_best_match function to the ngrams column

df['best_matching_bed_size'] = df['ngrams'].apply(lambda x: get_best_match(x, bed_size_key_words))

df['best_matching_room_style'] = df['ngrams'].apply(lambda x: get_best_match(x, room_style_key_words))

```

```python
import spacy
import pandas as pd

nlp = spacy.load("en_core_web_sm")

# sample data
data = {'room_type': ['Double City View King Bed', 'Deluxe Room Two Queen Beds', 'Twin City View Two Double Beds', 'Double Or Twin Two Queen Beds'], 
        'ngrams': [['Double', 'City', 'View', 'King', 'Bed'], ['Deluxe', 'Room', 'Two', 'Queen', 'Beds'], ['Twin', 'City', 'View', 'Two', 'Double', 'Beds'], ['Double', 'Or', 'Twin', 'Two', 'Queen', 'Beds']]}

df = pd.DataFrame(data)

# function to extract number from trigrams
def extract_bed_count(trigrams):
    for trigram in trigrams:
        if ("Bed" in trigram or "bed" in trigram or "Beds" in trigram or "beds" in trigram) and any(char.isdigit() for char in trigram):
            num = [int(s) for s in trigram.split() if s.isdigit()][0]
            return num
    return 1

# apply function to ngrams column
df['bed_count'] = df['ngrams'].apply(lambda x: extract_bed_count(x))

print(df)
```

```python
import pandas as pd

# define sample dataframe
df = pd.DataFrame({'room_type': ['Double Or Twin Two Queen Beds', 'Single', 'One King Bed', 'Triple Room with Three Beds']})

# define function to extract bed count from room type string
def extract_beds(room_type):
    # replace "One" with 1
    room_type = room_type.replace("One", "1")
    # replace "Two" with 2
    room_type = room_type.replace("Two", "2")
    # replace "Three" with 3
    room_type = room_type.replace("Three", "3")
    # return the updated string
    return room_type

# apply the extract_beds function to the 'room_type' column of the dataframe
df['room_type'] = df['room_type'].apply(extract_beds)

# print the updated dataframe
print(df)
```
