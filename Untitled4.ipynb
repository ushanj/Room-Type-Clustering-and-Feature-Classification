```python
# Define a function to tokenize the room types and extract bigrams and trigrams
# def get_ngrams(room_type):
#     tokens = room_type.split()
#     bigrams = [' '.join(tokens[i:i+2]) for i in range(len(tokens)-1)]
#     trigrams = [' '.join(tokens[i:i+3]) for i in range(len(tokens)-2)]
#     return tokens + bigrams + trigrams


def get_ngrams(room_type):
    # Replace occurrences of "digit space digit" with "digitdigit"
    room_type = room_type.replace(r'(\d) (\d)', r'\1\2')
    
    tokens = room_type.split()
    bigrams = [' '.join(tokens[i:i+2]) for i in range(len(tokens)-1)]
    trigrams = [' '.join(tokens[i:i+3]) for i in range(len(tokens)-2)]
    return tokens + bigrams + trigrams

# Apply the get_ngrams function to the room types column
df['ngrams'] = df['room_type'].apply(get_ngrams)
```

```python
import pandas as pd
import re

# Define a regular expression to match trigrams that end with "bed" or "beds"
regex = re.compile(r'\b\w+\s+\w+\s+(beds?|sofa bed)\b')

# Define a function to find the best matching bed size keyword for a given ngram
def get_best_match(ngram):
    matches = []
    for bed_size in bed_size_key_words:
        if bed_size in ngram:
            matches.append(bed_size)
    if len(matches) == 0:
        return None
    elif len(matches) == 1:
        return matches[0]
    else:
        # If there are multiple matches, return the longest match
        return max(matches, key=len)

# Find the best matching bed size keyword for each ngram and add it to a new column
df['bed_size'] = df['ngrams'].apply(lambda x: get_best_match(x))


```

```python

# Define a function to find the best matching key word for a list of ngrams
def get_best_match(ngrams, key_words):
    best_match = ''
    best_similarity = -1
    for ngram in ngrams:
        if ngram in model:
            for key_word in key_words:
                # Split key word into individual words
                key_word_tokens = key_word.split()
                # Check if all words in key word are present in ngram
                if all(token in ngram.split() for token in key_word_tokens):
                    similarity = model.similarity(ngram, key_word)
                    if similarity > best_similarity:
                        best_similarity = similarity
                        best_match = key_word
    return best_match



# Apply the get_best_match function to the ngrams column for bed sizes
#df['bed_size'] = df['ngrams'].apply(lambda x: get_best_match(x, bed_size_key_words))

# Apply the get_best_match function to the ngrams column for room styles
df['room_style'] = df['ngrams'].apply(lambda x: get_best_match(x, room_style_key_words))
```

```python
import spacy
nlp = spacy.load('en_core_web_sm')
```

```python

style_size_dict = {
    'Deluxe': 'King',
    'Corner': 'Queen',
    'Standard': 'Single',
    'Suit': 'Twin Beds'
    
}

def extract_style_size(row):
    room_type = row['hotelRoomType']
    room_style = row['room_style']
    bed_size = row['bed_size']
    
    if not room_style and not bed_size:
        for token in nlp(room_type):
            if token.text in style_size_dict:
                bed_size = style_size_dict[token.text]
                room_style = token.text
                break
    
    elif not room_style:
        for token in nlp(room_type):
            if token.text in style_size_dict:
                room_style = token.text
                break
                
    elif not bed_size:
        if room_style in style_size_dict:
            bed_size = style_size_dict[room_style]
    
    row['room_style'] = room_style
    row['bed_size'] = bed_size
    return row

df = df.apply(extract_style_size, axis=1)
```
